# Which country has won the most men's gold medals in singles badminton over the years?
# Sort the results alphabetically by the player's names.
# Total: (2 points)

```python
### 6.Which country has won the most men's gold medals in singles badminton over the years?
df_country = df_olympics.loc[(df_olympics["Gender"]=="Men")&(df_olympics["Medal"]=="Gold")&(df_olympics["Event"]=="singles")&(df_olympics["Sport"]=="Badminton")].reset_index(drop=True)
df_country = df_country.drop([1,1]).sort_values('NOC',ascending=True).reset_index(drop=True)
df_no1 = df_country.groupby("NOC")["Athlete"].count()
df_no1
```

```python
### 7.Sort the results alphabetically by the player's names.
df_n1 = df_country.sort_values('Athlete',ascending=True).reset_index(drop=True)
df_n1
```

# Which three countries have won the most medals in recent years (from 1984 to 2008)? (2 points)

```python
### 9.1 Filter the dataframe in recent years (from 1984 to 2008)
df_recent = df_olympics.loc[(1984<= df_olympics['Edition']) & (df_olympics['Edition'] <= 2008)]
df_recent
```

```python
### 9.2 Three countries have won the most medals
df_most_medals = df_recent.groupby("NOC")[['Medal']].count().sort_values('Medal',ascending=False).head(3)
df_most_medals
```

# Using groupby(), plot the total number of medals awarded at each of the Olympic games throughout history. (2 points)

```python
### 17.1 Using groupby() Event, get the total number of medals awarded at each of the Olympic games throughout history
df_games_medals = df_olympics.groupby("Event")[['Medal']].count().sort_values('Medal',ascending=False)
df_games_medals
```

```python
### 17.2 Using matplotlib, plot the total number of medals awarded at each of the Olympic games throughout history
df_games_medals.head(100).plot(kind='bar',color='steelblue',figsize=(20,10),fontsize=12)
plt.title("Total Number of Medals in Olympic Games",fontsize=16)
plt.xlabel("Event",fontsize=12)
plt.ylabel("Total number of Medals",fontsize=12)
plt.show()
```

```python
### 17.3 Using groupby() Edition, get the total number of medals awarded at each of the Olympic games throughout history
# df_games_medalsEd = df_olympics.groupby("Edition")[['Medal']].count().sort_values('Medal',ascending=False)
df_games_medalsEd = df_olympics.groupby("Edition")[['Medal']].count()
df_games_medalsEd
```

```python
### 17.4 Using matplotlib, plot the total number of medals awarded at each of the Olympic games throughout history
df_games_medalsEd.plot(kind='bar',color='steelblue',figsize=(20,10),fontsize=12) #color='#0099FF'   figsize=(20,10)
plt.title("Total Number of Medals in Olympic Games",fontsize=16)
plt.xlabel("Edition",fontsize=12)
plt.ylabel("Total number of Medals",fontsize=12)
plt.show()
```

# Create a list showing the total number of medals won for each country over the history of the Olympics.
# For each country, include the year of the first and most recent Olympic medal wins.
# Total: (3 points)

```python
### 18. Create a list showing the total number of medals won for each country over the history of the Olympics.
df_country_medals = df_olympics.groupby("NOC")[['Medal']].count().sort_values('Medal',ascending=False)
df_country_medals
```

```python
### 19. For each country, include the year of the first and most recent Olympic medal wins.
df_country_medals['First_Edition'] = df_olympics.groupby("NOC")[['Edition']].min()
df_country_medals['Recent_Edition'] = df_olympics.groupby("NOC")[['Edition']].max()
df_country_medals
```

# Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 using:
# - matplotlib
# - Seaborn
# Total: (3 points)


```python
### 21. Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008

### 21.1 The dataframe of medals achieved by the Chinese team (men and women) in Beijing 2008
df_China = df_olympics.loc[(df_olympics["NOC"]=="CHN")&(df_olympics["City"]=="Beijing")]
df_China
```

```python
### 21.2 The number of medals achieved by the Chinese team divided into different Gender in Beijing 2008
df_medals_gender = df_China.groupby("Gender")[['Medal']].count()

### 22. Using matplotlib,plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008
df_medals_gender.plot(kind='bar',figsize=(10,6),color='#1C4076')
plt.title("Total Number of Medals of Chinese Team in 2008 Olympic Games",fontsize =14)
plt.xlabel("Gender",fontsize=12)
plt.ylabel("Total Number of Medals",fontsize=12)
plt.show()
```

```python
### Using seaborn,plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008
df_mg = df_medals_gender.reset_index()
plt.rcParams["figure.figsize"] = (10,6)
sns.set_style("ticks")
sns.barplot(data=df_mg,x="Gender",y="Medal",hue='Gender',palette='Blues')
plt.show()
```
